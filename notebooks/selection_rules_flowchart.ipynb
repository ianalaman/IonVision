```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

fig, ax = plt.subplots(figsize=(20, 22))
ax.set_xlim(0, 20)
ax.set_ylim(0, 22)
ax.axis('off')

# Examples
example_yes = "Example E1:\n5s² S₁/₂ → 5p² P₁/₂\nParity changes → E1 allowed"
example_no = "Example M1/E2:\n5p² P₁/₂ → 5p² P₃/₂\nParity does NOT change → M1/E2 allowed"

# Helper function to add boxes
def add_box(x, y, text, example_text=None, example_color='darkred', width=5, height=2):
    ax.add_patch(Rectangle((x, y), width, height, fill=True, edgecolor='black', facecolor='#ffe5b4'))
    ax.text(x + width/2, y + height - 0.3, text, ha='center', va='top', fontsize=12, wrap=True)
    if example_text:
        ax.text(x + width/2, y + 0.3, example_text, ha='center', va='bottom',
                fontsize=10, color=example_color, style='italic')
    return width, height

# Helper function to draw short arrows
def add_short_arrow(x_start, y_start, x_end, y_end, label=None):
    ax.annotate('', xy=(x_end, y_end), xytext=(x_start, y_start),
                arrowprops=dict(facecolor='black', shrink=0.0, width=1, headwidth=8))
    if label:
        ax.text((x_start + x_end)/2, (y_start + y_end)/2 + 0.2, label, ha='center', va='center', fontsize=10)

# Boxes
boxes = {}
boxes['start'] = add_box(7, 20, 'Start: Given transition', f"{example_yes}\n{example_no}", example_color='darkgreen')
boxes['q1'] = add_box(7, 18, 'Q1: Parity change? (Even ↔ Odd)', 
                      "E1: parity changes → Yes\nM1/E2: parity same → No", example_color='darkgreen')
boxes['yes'] = add_box(2, 16, 'Consider E1, E3, M2', 
                       "Parity changed → E1 allowed,\nE3/M2 weaker")
boxes['no'] = add_box(12, 16, 'Consider M1, E2, M3', 
                      "Parity same → M1/E2/M3 allowed", example_color='darkgreen')
boxes['q2_yes'] = add_box(2, 14, 'Q2: ΔL?', 
                          "E1: ΔL = 0→1 → +1 → Dipole → E1 possible")
boxes['q2_no'] = add_box(12, 14, 'Q2: ΔL?', 
                         "M1: ΔL = 0 → allowed for M1\nE2: ΔL = 0 → quadrupole possible", example_color='darkgreen')
boxes['q3_yes'] = add_box(2, 12, 'Q3: Is ΔJ allowed?', 
                          "E1: J=1/2→1/2 ΔJ=0 → allowed")
boxes['q3_no'] = add_box(12, 12, 'Q3: Is ΔJ allowed?', 
                         "M1: J=1/2→3/2 ΔJ=+1 → allowed\nE2: ΔJ=+1 → allowed", example_color='darkgreen')
boxes['q4_yes'] = add_box(2, 10, 'Q4: ΔS = 0?', 
                          "E1: ΔS=0 → allowed")
boxes['q4_no'] = add_box(12, 10, 'Q4: ΔS = 0?', 
                         "M1/E2: ΔS=0 → allowed", example_color='darkgreen')
boxes['q5_yes'] = add_box(2, 7.5, 'Q5: Choose lowest-order allowed multipole?', 
                          "E1: lowest-order → pick E1")
boxes['q5_no'] = add_box(12, 7.5, 'Q5: Choose lowest-order allowed multipole?', 
                         "M1/E2: lowest-order → pick M1", example_color='darkgreen')
boxes['end_yes'] = add_box(2, 5, 'End: Transition classified', 
                           "5s² → 5p² → E1")
boxes['end_no'] = add_box(12, 5, 'End: Transition classified', 
                          "5p² → 5p² → M1 or E2", example_color='darkgreen')

# Arrows
add_short_arrow(9.5, 20, 9.5, 18 + boxes['q1'][1])                  # Start -> Q1
add_short_arrow(7, 18, 4.5, 16 + boxes['yes'][1], 'Yes')           # Q1 -> left
add_short_arrow(7, 18, 14.5, 16 + boxes['no'][1], 'No')            # Q1 -> right

add_short_arrow(4.5, 16, 4.5, 14 + boxes['q2_yes'][1])              # Yes -> Q2
add_short_arrow(14.5, 16, 14.5, 14 + boxes['q2_no'][1])            # No -> Q2

add_short_arrow(4.5, 14, 4.5, 12 + boxes['q3_yes'][1])             # Q2 -> Q3
add_short_arrow(14.5, 14, 14.5, 12 + boxes['q3_no'][1])            # Q2 -> Q3

add_short_arrow(4.5, 12, 4.5, 10 + boxes['q4_yes'][1])              # Q3 -> Q4
add_short_arrow(14.5, 12, 14.5, 10 + boxes['q4_no'][1])             # Q3 -> Q4

add_short_arrow(4.5, 10, 4.5, 7.5 + boxes['q5_yes'][1])            # Q4 -> Q5
add_short_arrow(14.5, 10, 14.5, 7.5 + boxes['q5_no'][1])           # Q4 -> Q5

add_short_arrow(4.5, 7.5, 4.5, 5 + boxes['end_yes'][1])            # Q5 -> End
add_short_arrow(14.5, 7.5, 14.5, 5 + boxes['end_no'][1])           # Q5 -> End

plt.show()
```
