```python
# Basic plotting and data setup
import matplotlib.pyplot as plt

# Define basic style
plt.rcParams.update({
    "figure.figsize": (8, 4),
    "font.size": 12,
    "axes.labelsize": 14,
})
```

```python
def plot_energy_levels(levels, title="Energy Level Diagram", unit="cm⁻¹"):
    fig, ax = plt.subplots()

    for i, (label, energy) in enumerate(levels):
        ax.hlines(energy, i, i + 0.5, colors="black")
        ax.text(i + 0.55, energy, label, va="center")

    ax.set_ylabel(f"Energy ({unit})")
    ax.set_xticks([])
    ax.set_title(title)
    ax.set_ylim(0, max(energy for _, energy in levels) + 2000)
    plt.tight_layout()
    plt.show()
```

```python
# Energy levels in cm⁻¹ from NIST
sr88_levels = [
    ("5s²S₁/₂", 0),
    ("5p²P₁/₂", 23737.1),
    ("4d²D₃/₂", 14504.4),
    ("5p²P₃/₂", 24356.8),
    ("4d²D₅/₂", 14835.6),
]

plot_energy_levels(sr88_levels, title="⁸⁸Sr⁺ Energy Levels")
```

```python
# Simplified Ca-43 structure, units: GHz (converted to cm⁻¹ ~ 1 GHz ≈ 0.033 cm⁻¹)
ca43_levels = [
    ("4S₁/₂ (F=4)", 0),
    ("4S₁/₂ (F=3)", 3.2 * 0.033),  # hyperfine splitting
    ("4P₁/₂", 25191),
    ("3D₅/₂", 13649),
]

plot_energy_levels(ca43_levels, title="⁴³Ca⁺ Energy Levels", unit="cm⁻¹")
```
